## Przygotowanie modeli do eksperymentów

```python
import adom_library as adom

adom.show_device()
```

```python
dataset = adom.get_CIFAR10_dataset()
classes = dataset.classes

epochs = 10
batch_size = 64
```

```python
model_vgg16 = adom.train_new_model(
    dataset,
    model_type="VGG16",
    epochs=epochs,
    batch_size=batch_size
)

adom.save_model(model_vgg16, "vgg16_cifar10", classes)
```

```python
model_alexnet = adom.train_new_model(
    dataset,
    model_type="AlexNet",
    epochs=epochs,
    batch_size=batch_size
)

adom.save_model(model_alexnet, "alexnet_cifar10", classes)
```

```python
model_googlenet = adom.train_new_model(
    dataset,
    model_type="GoogleNet",
    epochs=epochs,
    batch_size=batch_size
)

adom.save_model(model_googlenet, "googlenet_cifar10", classes)
```
